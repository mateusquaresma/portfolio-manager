```python
import pandas as pd
import numpy as np
import requests
import json
import time
```

```python
def get_quotes(company):
    url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=%s&apikey=K51ZJJVASBGYV9DD" % (company)
    company_quotes = []
    print('requesting ' + url)
    response = requests.get(url)
    if response.status_code == 200:
        content_data = json.loads(response.content)
        try:
            time_series = content_data['Time Series (Daily)']
            for i in time_series.items():
                date, quotes = i
                company_quotes.append(quotes['4. close'])

            company_quotes = np.array(company_quotes)
        except KeyError as e:
            company_quotes = np.zeros(100)
            print('KeyError on company %s: %s ' % (company, e))
    else:
        print('Error ' + response.status_code)
    
    time.sleep(1)
    
    return (company, company_quotes[::-1])
```
