```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import psycopg2
import seaborn as sbn
```

```python
query = """ 
    select dq.event_date, c.label, dq.close_value
    from company c
    inner join daily_quote dq on c.id = dq.company_id
    where c.label in (
        'CYRE3', 'BRKM5', 'RADL3', 'VVAR3', 'BBDC3', 'SBSP3', 'KROT3', 'EGIE3', 'ENBR3', 'NATU3')
    and (now() - dq.event_date) <= interval '3 months'
    order by dq.event_date;"""
```

```python
conn = psycopg2.connect(
            database="postgres",
            user="postgres",
            connect_timeout=3,
            options='-c statement_timeout=3000')
raw_df = pd.read_sql(query, conn)
conn.close()
```

```python
raw_df.columns = ['Data', 'Sigla', 'Valor_de_Fechamento']
raw_df.set_index(['Data', 'Sigla'], inplace=True)
```

```python
df = (raw_df.unstack()
      .replace({0: np.NaN})
      .fillna(method='ffill'))
df.head()
```

```python
columns = pd.Series(df.columns.levels[1].values)
```

```python
for label in df['Valor_de_Fechamento']:
    fig = plt.figure()
    company = df['Valor_de_Fechamento'][label]
    props = {
        'title': company.name
    }
    ax = fig.add_subplot(1, 1, 1)
    ax.label = company.name
    company.plot(ax=ax, style='b-')
    ax.set(**props)
```

```python
df.head()
percentages = [(13, 'CYRE3'), (13, 'BRKM5'), (12, 'RADL3'), (11, 'VVAR3'), (9, 'BBDC3'), (9, 'SBSP3'), (9, 'KROT3'), (8, 'EGIE3'),
 (8, 'ENBR3'), (8, 'NATU3')]


dict = {}
for (v, k) in percentages:
    dict[k] = [v] * len(df)

percentages_df = pd.DataFrame(dict)
percentages_df = percentages_df[['CYRE3', 'BRKM5', 'RADL3', 'VVAR3', 'BBDC3', 
                                 'SBSP3', 'KROT3', 'EGIE3', 'ENBR3', 'NATU3']]

virtual_portfolio = np.sum(np.multiply(df, np.multiply(percentages_df, 1/100)), axis=1)
virtual_portfolio.plot()
                    
```

```python
profit = virtual_portfolio.max() - virtual_portfolio.min()
print(profit)
print(profit/virtual_portfolio.min())
```
